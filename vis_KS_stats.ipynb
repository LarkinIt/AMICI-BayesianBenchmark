```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ks_df = pd.read_csv("results/linear_mRNA_self_reg_ks_data.csv", index_col=0)
shorthand = "mRNA"
ks_df
```

```python
plt.figure(figsize=(10,4.5), dpi=350)
sns.boxplot(data=ks_df, x="Param", hue="Method", y="KS", showfliers=True)
#plt.xticks(rotation=90);
plt.ylabel("Two-Way Kolmogorov-Smirnov Statistic")
plt.title(shorthand)
plt.axhline(y=0.05, linestyle="--", color="r", lw=2);
```
