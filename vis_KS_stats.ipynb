```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ks_df = pd.read_csv("Shp2_ks_data.csv", index_col=0)
ks_df
```

```python
plt.figure(figsize=(16,7), dpi=200)
sns.boxplot(data=ks_df, x="Param", hue="Method", y="KS", showfliers=True)
plt.xticks(rotation=90);
plt.ylabel("Two-Way Kolmogorov-Smirnov Statistic")
plt.axhline(y=0.05, linestyle="--", color="r", lw=2);
```
