```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ks_df = pd.read_csv("results/linear_mRNA_self_reg_ks_data.csv", index_col=0)
ks_df
```

```python
plt.figure(figsize=(12,4), dpi=350)
hueorder=["PT-MCMC", "SMC", "PMC"]
sns.boxplot(data=ks_df, x="Param", hue="Method", y="KS", showfliers=True, hue_order=hueorder)
#plt.xticks(rotation=90);
plt.ylabel("Two-Way Kolmogorov-Smirnov Statistic")
plt.xlabel("Parameter")
#plt.title(shorthand)
#plt.xticks(plt.gca().get_xticks(), ["$p_0$","$p_1$", "$p_2$","$p_3$","dummy"])
plt.axhline(y=0.05, linestyle="--", color="r", lw=2);
```
